```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
ride_distance = ctrl.Antecedent(np.arange(0, 51, 1), 'ride_distance')
traffic_condition = ctrl.Antecedent(np.arange(0, 101, 1), 'traffic_condition')
demand_level= ctrl.Antecedent(np.arange(0, 101, 1), 'demand_level')
weather_condition = ctrl.Antecedent(np.arange(0, 11, 1), 'weather_condition')
customer_rating = ctrl.Antecedent(np.arange(1, 6, 0.1), 'customer_rating')
ride_punctuality = ctrl.Antecedent(np.arange(0, 101, 1), 'ride_punctuality')
price = ctrl.Consequent(np.arange(0, 101, 1), 'price')
reward = ctrl.Consequent(np.arange(0, 11, 1), 'reward')
ride_distance['short'] = fuzz.trapmf(ride_distance.universe, [0, 0, 3, 5])
ride_distance['medium'] = fuzz.trimf(ride_distance.universe, [2, 5, 8])
ride_distance['long'] = fuzz.trimf(ride_distance.universe, [6, 13, 20])
ride_distance['very_long'] = fuzz.trapmf(ride_distance.universe, [15, 30, 50, 50])
for var in [traffic_condition, demand_level]:
    var['low'] = fuzz.trimf(var.universe, [0, 0, 30])
    var['medium'] = fuzz.trimf(var.universe, [20, 50, 80])
    var['high'] = fuzz.trimf(var.universe, [70, 100, 100])
weather_condition['good'] = fuzz.trimf(weather_condition.universe, [0, 0, 4])
weather_condition['moderate'] = fuzz.trimf(weather_condition.universe, [3, 5, 7])
weather_condition['bad'] = fuzz.trimf(weather_condition.universe, [6, 10, 10])
customer_rating['poor'] = fuzz.trimf(customer_rating.universe, [1, 1, 2.5])
customer_rating['average'] = fuzz.trimf(customer_rating.universe, [2, 3, 4])
customer_rating['good'] = fuzz.trimf(customer_rating.universe, [3.5, 5, 5])
ride_punctuality['late'] = fuzz.trimf(ride_punctuality.universe, [0, 0, 50])
ride_punctuality['on_time'] = fuzz.trimf(ride_punctuality.universe, [40, 60, 80])
ride_punctuality['early'] = fuzz.trimf(ride_punctuality.universe, [70, 100, 100])
price['low'] = fuzz.trimf(price.universe, [0, 0, 35])
price['medium'] = fuzz.trimf(price.universe, [30, 50, 70])
price['high'] = fuzz.trimf(price.universe, [65, 85, 100])
price['very_high'] = fuzz.trimf(price.universe, [80, 100, 100])
reward['none'] = fuzz.trimf(reward.universe, [0, 0, 2])
reward['few'] = fuzz.trimf(reward.universe, [1, 3, 5])
reward['moderate'] = fuzz.trimf(reward.universe, [4, 6, 8])
reward['high'] = fuzz.trimf(reward.universe, [7, 10, 10])
rules=[
  ctrl.Rule(ride_distance['short'] & traffic_condition['low'] & demand_level['low'], price['low']),
  ctrl.Rule(ride_distance['short'] & traffic_condition['medium'] & demand_level['high'], price['medium']),
  ctrl.Rule(ride_distance['medium'] & traffic_condition['high'] & demand_level['high'], price['high']),
  ctrl.Rule(ride_distance['long'] & traffic_condition['medium'] & weather_condition['good'], price['medium']),
  ctrl.Rule(ride_distance['long'] & traffic_condition['high'] & weather_condition['bad'], price['very_high']),
  ctrl.Rule(ride_distance['very_long'] & traffic_condition['high'] & demand_level['high'], price['very_high']),
  ctrl.Rule(ride_distance['medium'] & traffic_condition['low'] & demand_level['low'], price['medium']),
  ctrl.Rule(ride_distance['short'] & traffic_condition['high'] & weather_condition['bad'], price['high']),
  ctrl.Rule(ride_distance['very_long'] & weather_condition['bad'], price['very_high']),
  ctrl.Rule(ride_distance['medium'] & traffic_condition['medium'] & weather_condition['moderate'], price['medium']),
  ctrl.Rule(customer_rating['good'] & ride_punctuality['early'], reward['high']),                # Luật 11
  ctrl.Rule(customer_rating['average'] & ride_punctuality['on_time'], reward['moderate']),
  ctrl.Rule(customer_rating['poor'] & ride_punctuality['late'], reward['none']),
  ctrl.Rule(ride_distance['long'] & traffic_condition['high'] & ride_punctuality['on_time'], reward['high']),
  ctrl.Rule(ride_distance['medium'] & traffic_condition['medium'] & customer_rating['good'], reward['moderate']),
  ctrl.Rule(customer_rating['poor'] & ride_punctuality['late'], reward['none']),
  ctrl.Rule(ride_distance['long'] & traffic_condition['high']& ride_punctuality['late'], reward['few']),
  ctrl.Rule(ride_distance['medium'] & weather_condition['moderate'] & customer_rating['good'], reward['moderate'])
]
grab_system = ctrl.ControlSystem(rules)
grab_sim = ctrl.ControlSystemSimulation(grab_system)
# Giả sử: Chuyến đi dài 18km, trời mưa to (bad), khách đánh giá 4.8 sao, tài xế đến sớm.
grab_sim.input['ride_distance'] = 18
grab_sim.input['traffic_condition'] = 80
grab_sim.input['demand_level'] = 90
grab_sim.input['weather_condition'] = 9
grab_sim.input['customer_rating'] = 4.8
grab_sim.input['ride_punctuality'] = 100
grab_sim.compute()
print(f"KẾT QUẢ PHÂN TÍCH GRABBIKE")
print(f"Giá cước (thang 0-100): {grab_sim.output['price']:.2f}")
print(f"Điểm thưởng cụ thể: {grab_sim.output['reward']:.2f}")
price.view(sim=grab_sim)
reward.view(sim=grab_sim)
```

```python
#2.12
import numpy as np
import skfuzzy as fuzzy
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
store_rating = ctrl.Antecedent(np.arange(0, 5.1, 0.1), 'store_rating')
sales_volume = ctrl.Antecedent(np.arange(0, 101, 1), 'sales_volume')
profit_margin = ctrl.Antecedent(np.arange(0, 101, 1), 'profit_margin')
seasonal_event = ctrl.Antecedent(np.arange(0, 11, 1), 'seasonal_event')
competitor_discounts = ctrl.Antecedent(np.arange(0, 101, 1), 'competitor_discounts')
discount_rate = ctrl.Consequent(np.arange(0, 71, 1), 'discount_rate')
store_rating['low'] = fuzzy.trimf(store_rating.universe, [0, 0, 4.0])
store_rating['medium'] = fuzzy.trimf(store_rating.universe, [4.0, 4.25, 4.5])
store_rating['high'] = fuzzy.trimf(store_rating.universe, [4.5, 5.0, 5.0])
sales_volume['low'] = fuzzy.trimf(sales_volume.universe, [0, 0, 40])
sales_volume['medium'] = fuzzy.trimf(sales_volume.universe, [30, 50, 70])
sales_volume['high'] = fuzzy.trimf(sales_volume.universe, [60, 100, 100])
profit_margin['low'] = fuzzy.trimf(profit_margin.universe, [0, 0, 30])
profit_margin['medium'] = fuzzy.trimf(profit_margin.universe, [20, 50, 80])
profit_margin['high'] = fuzzy.trimf(profit_margin.universe, [70, 100, 100])
seasonal_event['none'] = fuzzy.trimf(seasonal_event.universe, [0, 0, 3])
seasonal_event['moderate'] = fuzzy.trimf(seasonal_event.universe, [2, 5, 8])
seasonal_event['high'] = fuzzy.trimf(seasonal_event.universe, [7, 10, 10])
competitor_discounts['low'] = fuzzy.trimf(competitor_discounts.universe, [0, 0, 30])
competitor_discounts['medium'] = fuzzy.trimf(competitor_discounts.universe, [20, 50, 80])
competitor_discounts['high'] = fuzzy.trimf(competitor_discounts.universe, [70, 100, 100])
discount_rate['very_low'] = fuzzy.trimf(discount_rate.universe, [0, 0, 5])
discount_rate['low'] = fuzzy.trimf(discount_rate.universe, [5, 10, 15])
discount_rate['medium'] = fuzzy.trimf(discount_rate.universe, [10, 20, 30])
discount_rate['high'] = fuzzy.trimf(discount_rate.universe, [20, 40, 50])
discount_rate['very_high'] = fuzzy.trimf(discount_rate.universe, [40, 70, 70])
rule1 = ctrl.Rule(store_rating['high'] & sales_volume['high'] & profit_margin['high'], discount_rate['very_low'])
rule2 = ctrl.Rule(store_rating['low'] & sales_volume['low'] & profit_margin['high'], discount_rate['high'])
rule3 = ctrl.Rule(seasonal_event['high'] & competitor_discounts['high'], discount_rate['very_high'])
rule4 = ctrl.Rule(store_rating['medium'] & sales_volume['medium'] & profit_margin['medium'], discount_rate['medium'])
rule5 = ctrl.Rule(competitor_discounts['low'] & profit_margin['low'] & sales_volume['high'], discount_rate['very_low'])
rule6 = ctrl.Rule(store_rating['low'] & seasonal_event['moderate'], discount_rate['medium'])
rule7 = ctrl.Rule(sales_volume['low'] & profit_margin['low'], discount_rate['very_high'])
discount_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7])
discount_sim = ctrl.ControlSystemSimulation(discount_ctrl)
discount_sim.input['store_rating'] = 4.3
discount_sim.input['sales_volume'] = 50
discount_sim.input['profit_margin'] = 15
discount_sim.input['seasonal_event'] = 9
discount_sim.input['competitor_discounts'] = 85
discount_sim.compute()
print(f"KẾT QUẢ TÍNH TOÁN")
print(f"Tỷ lệ chiết khấu đề xuất: {discount_sim.output['discount_rate']:.2f}%")
discount_rate.view(sim=discount_sim)
plt.show()
```

```python
#2.13
import numpy as np
import skfuzzy as fuzzy
from skfuzzy import control as ctrl
product_demand = ctrl.Antecedent(np.arange(0, 11, 1), 'Product Demand')
competitor_pressure = ctrl.Antecedent(np.arange(0, 11, 1), 'Competitor Pricing Pressure')
store_reputation = ctrl.Antecedent(np.arange(0, 11, 1), 'Store Reputation')
profit_margin = ctrl.Antecedent(np.arange(0, 11, 1), 'Profit Margin')
seasonal_demand = ctrl.Antecedent(np.arange(0, 11, 1), 'Seasonal Demand')
discount_percentage = ctrl.Consequent(np.arange(0, 71, 1), 'Discount Percentage')
for var in [product_demand, competitor_pressure, store_reputation, profit_margin, seasonal_demand]:
    var['Thấp'] = fuzzy.trimf(var.universe, [0, 0, 5])
    var['Trung bình'] = fuzzy.trimf(var.universe, [0, 5, 10])
    var['Cao'] = fuzzy.trimf(var.universe, [5, 10, 10])
discount_percentage['Rất thấp'] = fuzzy.trimf(discount_percentage.universe, [0, 2.5, 5])
discount_percentage['Thấp'] = fuzzy.trimf(discount_percentage.universe, [5, 7.5, 10])
discount_percentage['Trung bình'] = fuzzy.trimf(discount_percentage.universe, [10, 15, 20])
discount_percentage['Cao'] = fuzzy.trimf(discount_percentage.universe, [20, 30, 40])
discount_percentage['Rất cao'] = fuzzy.trimf(discount_percentage.universe, [40, 55, 70])
rule1 = ctrl.Rule(product_demand['Cao'] & competitor_pressure['Thấp'] & profit_margin['Thấp'],
                  discount_percentage['Rất thấp'])
rule2 = ctrl.Rule(product_demand['Thấp'] & competitor_pressure['Cao'] & profit_margin['Cao'],
                  discount_percentage['Cao'])
rule3 = ctrl.Rule(store_reputation['Cao'] & profit_margin['Trung bình'] & seasonal_demand['Cao'],
                  discount_percentage['Trung bình'])
rule4 = ctrl.Rule(competitor_pressure['Cao'] & seasonal_demand['Cao'] & profit_margin['Cao'],
                  discount_percentage['Rất cao'])
rule5 = ctrl.Rule(store_reputation['Thấp'] & product_demand['Trung bình'] & profit_margin['Thấp'],
                  discount_percentage['Trung bình'])
rule6 = ctrl.Rule(product_demand['Cao'] & seasonal_demand['Thấp'] & competitor_pressure['Thấp'],
                  discount_percentage['Rất thấp'])
rule7 = ctrl.Rule(profit_margin['Cao'] & competitor_pressure['Trung bình'] & seasonal_demand['Trung bình'],
                  discount_percentage['Trung bình'])
discount_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7])
discount_sim = ctrl.ControlSystemSimulation(discount_ctrl)
discount_sim.input['Product Demand'] = 9.0
discount_sim.input['Competitor Pricing Pressure'] = 5.0
discount_sim.input['Store Reputation'] = 4.2
discount_sim.input['Profit Margin'] = 9.0
discount_sim.input['Seasonal Demand'] = 9.5
discount_sim.compute()
print(f"Mức giảm giá gợi ý: {discount_sim.output['Discount Percentage']:.2f}%")
discount_percentage.view(sim=discount_sim)
```

```python
#2.14
import numpy as np
import skfuzzy as fuzzy
from skfuzzy import control as ctrl
order_density = ctrl.Antecedent(np.arange(0, 11, 1), 'order_density')
delivery_urgency = ctrl.Antecedent(np.arange(0, 11, 1), 'delivery_urgency')
driver_load = ctrl.Antecedent(np.arange(0, 11, 1), 'driver_load')
traffic_conditions = ctrl.Antecedent(np.arange(0, 11, 1), 'traffic_conditions')
profit_per_delivery = ctrl.Antecedent(np.arange(0, 11, 1), 'profit_per_delivery')
orders_to_combine = ctrl.Consequent(np.arange(0, 11, 1), 'orders_to_combine')
delivery_priority = ctrl.Consequent(np.arange(0, 11, 1), 'delivery_priority')
names = ['low', 'medium', 'high']
for var in [order_density, delivery_urgency, driver_load, traffic_conditions, profit_per_delivery, delivery_priority]:
    var.automf(3, names=names)
orders_to_combine.automf(3, names=['few', 'some', 'many'])
rules = [
    ctrl.Rule(order_density['high'] & driver_load['low'] & traffic_conditions['low'], orders_to_combine['many']),
    ctrl.Rule(order_density['medium'] & traffic_conditions['high'] & delivery_urgency['medium'], orders_to_combine['some']),
    ctrl.Rule(driver_load['high'] & order_density['high'] & profit_per_delivery['medium'], orders_to_combine['some']),
    ctrl.Rule(order_density['low'] & delivery_urgency['high'] & traffic_conditions['medium'], orders_to_combine['some']),
    ctrl.Rule(profit_per_delivery['high'] & delivery_urgency['high'] & traffic_conditions['high'], orders_to_combine['some']),
    ctrl.Rule(delivery_urgency['high'] & profit_per_delivery['high'], delivery_priority['high']),
    ctrl.Rule(delivery_urgency['medium'] & traffic_conditions['medium'], delivery_priority['medium']),
    ctrl.Rule(delivery_urgency['low'] & order_density['high'] & profit_per_delivery['low'], delivery_priority['low']),
    ctrl.Rule(order_density['low'], orders_to_combine['few']),
    ctrl.Rule(delivery_urgency['low'], delivery_priority['low'])
]
delivery_ctrl = ctrl.ControlSystem(rules)
delivery_sim = ctrl.ControlSystemSimulation(delivery_ctrl)
delivery_sim.input['order_density'] = 8.5
delivery_sim.input['delivery_urgency'] = 5.0
delivery_sim.input['driver_load'] = 2.0
delivery_sim.input['traffic_conditions'] = 3.0
delivery_sim.input['profit_per_delivery'] = 6.0
try:
    delivery_sim.compute()
    print(f"Kết quả giải mờ:")
    print(f"- Số lượng đơn hàng cần kết hợp: {delivery_sim.output['orders_to_combine']:.2f}")
    print(f"- Mức độ ưu tiên giao hàng: {delivery_sim.output['delivery_priority']:.2f}")
    orders_to_combine.view(sim=delivery_sim)
    delivery_priority.view(sim=delivery_sim)
except KeyError as e:
    print(f"Input hiện tại chưa khớp với luật nào.")
```
